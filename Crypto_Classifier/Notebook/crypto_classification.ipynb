# Feature Engineering & Data Preprocessing

 While I was researching about the address of different cryptocurrencies, I found that the address of the cryptocurrencies are of different lengths like address of bitcoin usually starts with 1 or 3 and is of length 34, address of ethereum is of length 42 and starts with 0x, address of dogecoin is of length 34 and starts with D, address of dash is of length 34 and starts with X. So, I will be creating a new column 'address_length' which will contain the length of the address and 'address_prefix' which will contain the starting character of the address.

```python
# creating the columns which I mentioned above

df['address_length'] = df['Address'].apply(len)
df['address_prefix'] = df['Address'].str[:2]
```

```python
# For the next feature engineering steps I will be using the 'address_prefix' column. So, I will be converting the 'address_prefix' column to numerical values using the LabelEncoder

from sklearn.preprocessing import LabelEncoder
le_for_x = LabelEncoder()
df_without_duplicates['address_prefix'] = le_for_x.fit_transform(df_without_duplicates['address_prefix'])

# Now, I will be dropping the 'Address' column as it is not required for the further steps
df_without_duplicates.drop('Address', axis=1, inplace=True)

df_without_duplicates.head()
```

# Train Test Split and Model Selection

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X = df_without_duplicates.drop('crypto_name', axis=1)
y = df_without_duplicates['crypto_name']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.22, random_state=10)
```

```python
# Scaling the data

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```

```python
# Now let's decide which model to choose for predicting the crypto's name. 
# For this I'll be creating one function to check the accuracy of each given model

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

def evaluate_models(models, X_train, y_train, X_test, y_test, y_encoded_train, y_encoded_test):
    """
    Evaluate multiple models on a given dataset.

    Parameters:
    - models: List of tuples containing (model_name, model_instance)
    - X_train, y_train: Training data and labels
    - X_test, y_test: Testing data and labels

    Prints the accuracy for each model on the testing set.
    """
    for name, model in models:
        if name == "XGBoost":
            model.fit(X_train, y_encoded_train)
            y_pred = model.predict(X_test)
            accuracy = accuracy_score(y_encoded_test, y_pred)
            print(f"{name}: Accuracy = {accuracy:.4f}")
        else:
            model.fit(X_train, y_train)
            y_pred = model.predict(X_test)
            accuracy = accuracy_score(y_test, y_pred)
            print(f"{name}: Accuracy = {accuracy:.4f}")

models = [
    ("Random Forest", RandomForestClassifier()),
    ("Gradient Boosting", GradientBoostingClassifier()),
    ("Support Vector Machine", SVC()),
    ("Decision Tree", DecisionTreeClassifier()),
    ("K-Nearest Neighbors", KNeighborsClassifier()),
    ("Gaussian Naive Bayes", GaussianNB()),
    ("Logistic Regression", LogisticRegression(solver='liblinear')),
    ("XGBoost", XGBClassifier()),
]

xgb_encoder = LabelEncoder()
y_encoded_train = xgb_encoder.fit_transform(y_train)
y_encoded_test = xgb_encoder.transform(y_test)

evaluate_models(models, X_train_scaled, y_train, X_test_scaled, y_test, y_encoded_train, y_encoded_test)
```

From the above results we can see that the Random Forest, XGBoost and Gradient Boosting algorithms are giving maximum accuracies so, we will try to do hyperparameter tuning on these algorithms and see what gives the best result after tuning them. The reason why I am ignoring Decision Tress as this algorithm has higher possibility of Overfitiing.
