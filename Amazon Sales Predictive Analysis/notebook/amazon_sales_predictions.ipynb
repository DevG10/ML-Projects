```python
# From above we can observe that every columns are by-default is of the type 'object' but we should convert the columns which can be casted as type 'int' or 'float' so we will now typecast every columns which is suitable to convert. Also, here for conversion we have to first remove the '%' and '₹' as they can't be directly converted to float.

df['discounted_price'] = df['discounted_price'].str.replace('₹', '').str.replace(',', '')
df['actual_price'] = df['actual_price'].str.replace('₹', '').str.replace(',', '')
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '')
df['rating_count'] = df['rating_count'].str.replace(',', '')
```

```python
df.head()
```

```python
# Before converting the columns we have to check and go thorugh the dataset manually at once as there might be some values which can't be typecasted so we will create one function which will show us the non castable value along with it's index which will be returned in the form of the dictionary.
columns_to_convert = ["discounted_price", "actual_price", "discount_percentage", "rating", "rating_count"]

def find_non_castable_symbols(df):
    non_castable_symbols = {}
    for column in df.columns:
        non_castable_symbols[column] = []
        for index, value in df[column].items():
            try:
                # Attempt to cast the value to a numeric type
                float(value)
            except (ValueError, TypeError):
                # If casting fails, consider it a non-castable symbol
                non_castable_symbols[column].append((index, value))
    return non_castable_symbols

non_castable_symbols = find_non_castable_symbols(df[columns_to_convert])

# Print non-castable symbols for each column
for column, symbols in non_castable_symbols.items():
    print(f"Non-castable symbols in '{column}': {symbols}")
  
```

```python
# From the above code we can clearly see that the symbol '|' is non castable so we will repalce it at the row number 1279 of the rating column.

df.loc[1279, 'rating'] = 1

# Here we cannot directly repalce the value with mean as currently this data type is not been typecasted yet because of this special symbol so temporarily we will mark it to 1 not 0 as the rating is minimum 1
```

```python
# Now all columns are ready to be converted to the proper datatype.
columns_to_convert = ["discounted_price", "actual_price", "discount_percentage", "rating", "rating_count"]

for cols in columns_to_convert:
    df[cols] = df[cols].astype(float)
```

```python
# Updating our temporary assigned variable with the mean of the rating column

df.loc[1279, 'rating'] = round(df['rating'].mean())
```

```python
category_counts = df['category'].value_counts()
top_categories = category_counts.head(5)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x = top_categories.index, y = top_categories.values, palette = 'viridis')
plt.xticks(rotation = 90)
plt.xlabel('Category')
plt.ylabel('Number of Items Sold')
plt.title('Most Number of Categories Sold')
plt.show()
```

```python
# Categories with most positive and most negative ratings


average_reviews = df.groupby('category')['rating'].mean()

# Number of top and bottom categories to display
N = 5

# Sort categories by average review scores in descending order
sorted_categories = average_reviews.sort_values(ascending=False)

# Select the top N categories with highest ratings
top_categories = sorted_categories.head(N)

# Select the bottom N categories with lowest ratings
bottom_categories = sorted_categories.tail(N)

# Combine top and bottom categories
selected_categories = pd.concat([top_categories, bottom_categories])

```

```python
#Visulaizing the categories with best and worst ratings

plt.figure(figsize=(10, 6))
sns.barplot(x=selected_categories.index, y=selected_categories.values, palette='coolwarm')
plt.xticks(rotation = 90)  # Rotate x-axis labels for better visibility
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.title(f'Top {N} and Bottom {N} Categories by Average Rating')
```

```python
# Since the category names in our dataset are very huge we will try to shorten it keeping their meaning same.

data = {'category' : [category for category in df['category']]}

data_df = pd.DataFrame(data)

def create_short_names(category):
    subcategory = category.split('|')
    short_category = [sub[:3] for sub in subcategory]
    return '|'.join(short_category)
    
df['short_category'] = df['category'].apply(create_short_names)
```
